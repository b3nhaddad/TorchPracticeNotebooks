```python
import numpy as np
import cv2 as cv
import polars as pl
import os

img = cv.imread('something.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
```

```python
#img
kernal = np.ones((3,3),np.uint8)
kernal = kernal / 3
kernalWidth, kernalHeight = kernal.shape
```

```python
height,width,channels = img.shape
print(height,width,channels)
```

```python
sizeX = round(height/kernalWidth)
sizeY = round(width/kernalHeight)
print(sizeX,sizeY)
print(f"Times kernal should be slid on the X axis {sizeX}")
print(f"Times kernal should be slid on the Y axis {sizeY}")
```

```python
def blur(img,kernal):
  height,width,channels = img.shape
for i in range(0, width, 5):
  for j in range(0, height, 5):

    x_end = min(i + 5, width) #ensures the slices isn't wider than the image itselft
    y_end = min(j + 5, height) #ensures the slices isn't taller than the image itselft
    segment = img[j:y_end, i:x_end]
    outline = segment @ kernal
    max_value = np.max(outline)
    outline = np.where(outline == max_value, max_value, 0)
    img[j:y_end, i:x_end] = outline
```

```python
cv.imwrite("outlinedImage.jpg",img)
```
